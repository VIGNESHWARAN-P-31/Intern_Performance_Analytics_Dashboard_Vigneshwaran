```python
import pandas as pd

df = pd.read_csv("INTERN_DATASET.csv")

df.head()
```

Statistical Analysis

Check dataset summary

```python
df.describe()
```

Check dataset information

```python
df.info()
```

Check unique domains

```python
df["Domain"].value_counts()
```

Check performance level count

```python
df["Performance_Level"].value_counts()
```

Identify Trends and Patterns

Find average mentor rating by domain

```python
df.groupby("Domain")["Mentor_Rating"].mean()
```

Find average attendance by domain

```python
df.groupby("Domain")["Attendance_Percentage"].mean()
```

Find task completion rate

```python
df["Completion_Rate"] = df["Tasks_Completed"] / df["Tasks_Assigned"]
df.head()
```

Create Visualizations

Import visualization libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Domain Distribution

```python
sns.countplot(x="Domain", data=df)
plt.xticks(rotation=45)
plt.title("Intern Distribution by Domain")
plt.show()
```

Performance Level Distribution

```python
sns.countplot(x="Performance_Level", data=df)
plt.title("Intern Performance Level Distribution")
plt.show()
```

Mentor Rating by Domain

```python
sns.barplot(x="Domain", y="Mentor_Rating", data=df)
plt.xticks(rotation=45)
plt.title("Average Mentor Rating by Domain")
plt.show()
```

```python
sns.barplot(x="Performance_Level",
            y="Attendance_Percentage",
            data=df)

plt.title("Average Attendance by Performance Level")
plt.show()
```
